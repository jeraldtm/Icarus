```python
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import hvplot.xarray
```

```python
lox = 10.0
hix = 14.0
stepx = 0.05
loy = 11.0
hiy = 15.0
stepy = 0.05
```

```python
xpts = np.arange(lox, hix + stepx, stepx)
ypts = np.arange(loy, hiy + stepy, stepy)
```

```python
phi0 = pd.read_csv('probe_wire_side_1.55mm_shim_fieldRaster_2019-02-05_1.csv', comment='#')
```

```python
phi90 = pd.read_csv('probe_wire_side_1.55mm_shim_fieldRaster_2019-02-05_2.csv', comment='#')
```

```python
phi0_dv = {}
phi90_dv = {}
for c in phi0.columns[:]:
    phi0_dv[c] = (["Ycoord", "Xcoord"], phi0[c].values.reshape(xpts.size, ypts.size))
    phi90_dv[c] = (["Ycoord", "Xcoord"], phi90[c].values.reshape(xpts.size, ypts.size))
```

```python
coords = {'Xcoord': ('Xcoord', xpts), 'Ycoord': ('Ycoord', ypts)}
```

```python
phi0_ds = xr.Dataset(phi0_dv, coords)
phi90_ds = xr.Dataset(phi90_dv, coords)
```

```python
phi0_ds.X.hvplot()
```

```python
phi0_ds.Yfield_avg.hvplot()
```

```python
phi0_ds.Zfield_avg.hvplot()
```

```python
phi0_ds.Xfield_avg.hvplot()
```

```python
Zfield_mod = (phi90_ds.Zfield_avg**2 + phi0_ds.Zfield_avg**2)
```

```python
Zfield_mod.hvplot(title = "Combined 0 and 90 Z field")
```

```python
Zfield_mod.where(Zfield_mod == Zfield_mod.min(), drop=True)
```

```python
Yfield_mod = (phi90_ds.Xfield_avg**2 + phi0_ds.Yfield_avg**2)
```

```python
Yfield_mod.hvplot(title="Combined 0 and 90 Y field")
```

```python
Yfield_mod.where(Yfield_mod == Yfield_mod.max(), drop=True)
```

# From these, seems like the best point is $(13.5,12.9)$ with pretty strong corroboration beetween $Y$ maximum and $Z$ minimum field strength points

```python
Xfield_mod = (phi90_ds.Yfield_avg**2 + phi0_ds.Xfield_avg**2)
Xfield_mod.hvplot(title="Combined 0 and 90 X field")
```

```python
Xfield_mod.where(Xfield_mod == Xfield_mod.min(), drop=True)
```
